```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets
census_data_path = "/Users/beauweisman/Downloads/census_data.csv"
county_codes_path = "/Users/beauweisman/Downloads/county_codes.csv"
census_data = pd.read_csv(census_data_path)
county_codes = pd.read_csv(county_codes_path)

# Step 1: Find Davidson County Code
davidson_code = county_codes[county_codes['County'].str.contains("Davidson", case=False, na=False)]['County cod'].iloc[0]
print(f"Davidson County Code: {davidson_code}")

# Step 2: Filter Census Data for Davidson County and Relevant Years
davidson_census = census_data[
    (census_data['county'] == davidson_code) & 
    (census_data['year'].isin([1820, 1830, 1840]))
]

# Sum the total population for each year
population_by_year = davidson_census.groupby('year')['ntotal'].sum()
print("Population by Year:")
print(population_by_year)

# Step 3: Plot the Population Change
plt.figure(figsize=(8, 5))
population_by_year.plot(kind='line', marker='o', title='Population of Davidson County (1820-1840)')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.grid(True)
plt.show()
```
